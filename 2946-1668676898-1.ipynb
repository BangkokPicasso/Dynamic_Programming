exercise 4 求取最大公因數
利用轉轉相除法

```python
def gcd(m,n):
  if n == 0:   #終止條件      6 3 
    return m
  else :
    return gcd(n ,m % n)
  
m = int(input("輸入 m:"))
n = int(input("輸入 n:"))
print("最大公因數", gcd(m,n))
```

exercise 5 畫圖

```python
d = {}                      #儲存變數
for p in range(5, 21):             #算出利潤式
   profit = (20 - p) * (p - 5)
   d[p] = profit
print(d)

import matplotlib.pyplot as plt         #引入套件
#from google.colab import files

l1 = d.items()
l2 = sorted(l1)
x, y = zip(*l2)  #把排序好的資料分為 x , y
print(x,y)
plt.xlim(5,20)    #範圍
plt.ylim(0,60)
plt.xticks([5, 10, 15, 20], color = 'black') #刻度
plt.yticks([0,10,20, 30, 40, 50, 60] ,color = 'black')
plt.plot(x, y, color = 'r', label = 'profit line') #線條
plt.legend(loc = 'upper left')              #label放置的位置
plt.title("pricing decision ", color = 'green', fontsize = 14)
plt.xlabel('price', color = "b")
plt.ylabel('profit', color = "b")

#畫出最大值兩點
ax = plt.gca()         
ax.yaxis.grid(True)        #讓圖示有網格線
ax.plot([12], [56], 'o', color = "black")
plt.text(10, 56, "(12, 56)", fontsize = 8, color = "black")
ax.plot([13], [56], 'o', color = "black")
plt.text(14, 56, "(13, 56)", fontsize = 8, color = "black")
plt.tight_layout()
#plt.savefig("price.png")
#files.download("price.png")
plt.show()
```
