```python
cost_dict = {}
cost_dict[1] = {1:700,2:600,3:500}
cost_dict[2] = {1:450,2:300,3:400}
cost_dict[3] = {1:200,2:150,3:100}
all_min = {} #放各期最佳決策的字典
for i in cost_dict:#期數
  min_value = 10000000  #懲罰值
  min_key = 0
  for j in cost_dict[i]: #決策
    if min_value>cost_dict[i][j]:
      min_value=cost_dict[i][j]
      min_key=(i,j)
  all_min[min_key]=min_value
print(all_min)
```

```python
#建立一個函式讓其可以加總到 輸入的數值
def sum1(n):
  sum = 0
  for i in range(n+1):
    sum += i
  return sum

sum1(15)
```

```python
# 階乘 
def factorial(n):         #4! = 1*2*3*4
  if n==0 or n==1 :       # 4
    return 1
  else :
    return n*factorial(n-1)   # 4*f(3)  f(3) = 3 * f(2) f(2) = 2*f(1)

factorial(4)
```

```python
# 費式數列 
def fibonacci(n):
  if n ==0:
    return 0
  elif n == 1:
    return 1
  else :
    return fibonacci(n-1) + fibonacci(n-2)

fibonacci(7)
```

```python

data = [89, 34, 23, 78, 67, 100, 66, 29, 79, 55, 78, 88, 92, 96, 96, 23]    

def BubbleSort(data):
  n = len(data)   # 16
  while n > 1:                          # 16 15 14 13 12 11 1
    n = n - 1
    for i in range(n):                      # 0 1 2 3... 14       #
      if data[i] > data[i+1]: 
        #print(data[i] ,data[i+1] )
        data[i], data[i+1] = data[i+1], data[i]
        #print(data[i] ,data[i+1] )
        #print(data)
  return data

print(BubbleSort(data))
```

```python
import numpy as np # 引入NumPy
import matplotlib.pyplot as plt # 引入matplotlib的函數

x = np.arange(20) # x軸的值
y1 = 3 * x + 1  # y軸的值
y2 = -3 * x + 20  # y軸的值

plt.title("add legend") # 圖的標題
plt.xlabel("x axis") # x軸的名稱
plt.ylabel("y axis") # y軸的名稱
plt.plot(x,y1, color = "red", label = "y1") # 繪製x,y1的圖
plt.plot(x,y2, ls = "--", label = "y2") # 繪製x,y2的圖
plt.legend(loc = 0, prop={'size': "x-large"})
plt.show() # 顯現圖形
```
